```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame (adjust the file path as needed)
df = pd.read_csv(r'C:\Users\HP\Downloads\AlexTheAnalyst2.github.io-main\b68782efa49a16edaf07dc2cdaa855ea-0c794a9717f18b094eabab2cd6a6b9a226903577\b68782efa49a16edaf07dc2cdaa855ea-0c794a9717f18b094eabab2cd6a6b9a226903577\movies.csv')

# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])

# Generate the correlation matrix for the numeric features in the DataFrame
correlation_matrix = numeric_df.corr()

# Create the heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Set the title for the heatmap
plt.title("Correlation Matrix for Numeric Features")

# Show the plot
plt.show()
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Year")

plt.ylabel("Year")

plt.show()
```
